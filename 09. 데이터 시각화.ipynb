```r
# 3종류 dataset의 선 그래프
abc <- c(260, 300, 250,280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
# 같은 cell안에서 사용한다!
# 계속 보면서 해야함 그래서 note사용중인거임!
#plot 빼 놓고 바깥에 있는 함수들은 아무데나 같은셀 안에서 사용할수 있다.

par(family = "AppleGothic")
plot(abc,
    ylim = c(0, 400),#그래프의 범위가 서로 다름 이걸 잘봐야함. 내가 판단해야 하는것
    type = "o",
    col  = "red",
    axes = F, #x,y 죄표 Frame 제거(축이 두개니까 s)
    ann  = F #x,y의 Label제거
    )

# x축과 x축의 라벨을 설정(축만드는 함수 = axis)
axis(1, at=1:5, lab=c("서울","대전","대구","광주","원주"))

# y축과 y축의 라벨을 설정
axis(2, ylim=c(0:400))

#main title 지정 : plot에서 쓸수 있으나 같은셀 안에서 plot에 나와서 쓸수 있다
# 따로 쓰면 컬러값을 줄수 있음
title(main = "과일 판매량", col.main="white")

#X축 title(label)지정
title(xlab = "지역", col.lab = "yellow") 

#y축 title(label)지정
title(ylab = "판매량", col.lab = "yellow")


#선그리기 
#함수로 def그리기
lines(def,
     type = "o",
      pch = 22,
      col = "green",
      lty = 2
     )

#선그리기 함수로 ghi 그리기
lines(ghi,
      type = "o",
      pch  = 22,
      col  = "skyblue",
      lty  = 2
     )

#범례 (legend)# 먼저 사용할 위치를 정한다 data값으로 위치를 정한다.(index값이랑 데이터값)
legend(4, 400, # x,y 좌표
       c("야구장","축구장","해변가"),
       col = c("red","green","skyblue"),
       pch = c(21,22,22),
       lty = c(1,2,2),
       cex = 0.8 # 크기조정 기본값은 '1'
      )
```

## 막대 그래프

- 꼭 데이터 값 을 찍어보고 꾸미기!
- 막대 그래프는 값을 몰라서 ~~ 로 바꿔줘야함.
- 클라이언트한테 보여줄때 많이 사용
- 데이터 프레임 사용 불가
- 메트릭스 좋아함

```r
x <- 1:6 #x는 벡터임.
barplot(x,
       main = "1~6",
        xlab = "Sales",
        ylab = "Amounts",
        names = c("A","B","C","D","E","F"), #백터에서 이름주는게 names니까 사용
        col = c("pink",rep("skyblue",5))
       )
```

```r
# 복잡한 형태의 막대 그래프
# 3행 2열의 matrix생성

x <- matrix(1:6, 3, 2) # matrix로 3행2열을 만듦
x
barplot(x,
       ylim = c(0,16) # 처음 데이터보다 y축이 작아서 키워줌
       )
```

```r
# 누적 막대 그래흐 분리후 열별 label붙이기
#기본적으로 legend를 가지고 있다.
par(family = "AppleGothic")

#legend 쓸라면 rownames 잇어야함.
#legend 만들어줬어! 이제 불러오면됨!
rownames(x) <- c('1월', '2월', '3월')

barplot(x,
       #**beside : 열별로 나옴! (꼭 기억)
       beside = T, #누적부분 해체 
        ylim = c(0,8),
        names = c("korea", "America"),
        #names = c("K1","K2","k3","A1","A2","A3")
        main = "수출량",
        xlab = "국가",
        ylab = "빈도수",
        col = c("pink","green","skyblue"),
        legend.text = T
       )
```

# 3D Pie Chart
- Anaconda 에서 plotrix 다운

```r
library(plotrix)
```

```r
par(family = "AppleGothic")
sales <- c(210, 110, 400, 550, 600, 130)
week <- c("월","화","수","목","금","토") # 라벨달꺼 미리 정해주고
ratio <- sales / sum(sales) * 100 #퍼센트를 넣어주고싶다.
# 라벨구성 어떻게 하는지 알아두자!
label <- paste(week,"\n",ratio,"%")

pie3D(sales,
    main = "영업실적 파이차트",
    labels = label,
   # col = rainbow(length(sales)) 3D 는기본값이 레인보우임
    explode = 0.05,
    labelcex = 0.8
   )
```
