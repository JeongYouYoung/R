## 기본 통계량으로 분석해 보기
- summary 에서 보여주는게 기본 통계량

```r
#EX )다이어트약

# 밀가루 사용
flour <- c(3,-2,-1,0,1,-2)

#다이어트약
diet <-c(-4,1,-3,-5,-2,-8)
```

```r
total <- c(flour,diet)
total
```

```r
# 기본통계량
summary(total)
```

```r
hist(total)
```

```r
boxplot(total)
```

```r
boxplot(flour, diet)
```

```r
#상관계수 비교
cor(flour, diet)
```

```r
# 상관계수 언제 쓰나
# 순서가 엄청 중요하다!
x <- c(1,3,9)
y <- c(10,20,30)
```

```r
par(mfcol = c(1,2))
plot(x, type = "l")
plot(y, type = "l")
```

```r
cafe <- read.csv("Rdata/coffee.csv")
head(cafe)
```

```r
#커피의 빈도수
table(cafe$Coffee)
```

```r
# 범위 빈도수
#table(cut)
Coffee_Range <- seq(0,60,10) #범위
#right = F 는 (를 [로 바꿔서
table(cut(cafe$Coffee,breaks = Coffee_Range, right = F))
# ( => 포함 안한다.
# [ => 포함 한다.
```

```r
# 줄기 - 잎 그림
# 진짜 간단하게 보여주는 시각화

stem(cafe$Coffee)
```

```r
# 자주 판매된다. = 커피와 판매 개수를 비교
# 변동 계수 
head(cafe)
```

```r
cafe.mean <- aggregate(Coffee ~ Day.of.Week, cafe,mean )
cafe.sd <- aggregate(Coffee ~ Day.of.Week, cafe,sd )
cafe.mean
cafe.sd
```

```r
cafe.cv <- cafe.sd$Coffee/cafe.mean$Coffee
cafe.cv
```
