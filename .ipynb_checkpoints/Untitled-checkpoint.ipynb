```r
a <- read.csv("Rdata/서울시_우리마을가게_상권분석서비스_2020년.csv",fileEncoding = "euc-kr")
head(a)
```

```r
subset((a,(서비스_업종_코드_명=="호프-간이주점"&상권_구분_코드_명=='관광특구')),기준_분기_코드) # 관광특구 내 호프-간이주점의 2021년 데이터셋[,-c(3,5,7,8,11:79)]
```

```r
ncol(a)
unique(a$상권_구분_코드_명)
```

```r
#subset(a, 서비스_업종_코드_명=="호프-간이주점")
```

```r
b골 <- subset(a, (서비스_업종_코드_명=='호프-간이주점')&(상권_구분_코드_명 == '골목상권'))
b발 <- subset(a, (서비스_업종_코드_명=='호프-간이주점')&(상권_구분_코드_명 == '발달상권'))
b전 <- subset(a, (서비스_업종_코드_명=='호프-간이주점')&(상권_구분_코드_명 == '전통시장'))
b관 <- subset(a, (서비스_업종_코드_명=='호프-간이주점')&(상권_구분_코드_명 == '관광특구'))
head(b골)
head(b발)
head(b전)
head(b관)
```

```r
nrow(b골)
nrow(b발)
nrow(b전)
nrow(b관)
```

```r
bb골 <- aggregate(cbind(분기당_매출_금액,점포수)~기준_분기_코드,b골, sum)
bb발 <- aggregate(cbind(분기당_매출_금액,점포수)~기준_분기_코드,b발, sum)
bb전 <- aggregate(cbind(분기당_매출_금액,점포수)~기준_분기_코드,b전, sum)
bb관 <- aggregate(cbind(분기당_매출_금액,점포수)~기준_분기_코드,b관, sum)
```

```r
hof <- rbind(bb골, bb발, bb전, bb관)
hof
```
