### 2행 2열 차트 그리기
- mfrow
- (화면을 4등분 해서 사용하겠다. 같은 셀안에서 사용)

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해보기

par(mfrow = c(2,2))
plot(price, type ="l")
plot(price, type ="o")
plot(price, type ="h")
plot(price, type ="s")
```

```r
# 2행 2열 차트그리기 : pch 속성으로 구분해보기

par(mfrow = c(2,2))
plot(price, type ="o", pch=5)
plot(price, type ="o", pch=15)
plot(price, type ="o",pch=15, col = "blue")
plot(price, type ="o",pch=c(rep(15,8),20,20)
                     ,col = "orange"
                     ,cex = c(rep(1,8),3,1)
    )
```

### ***** 중첩 자료 시각화

- sunflowerplot()

```r
x <- c(1,2,3,4,2,4)
y <- rep(2,6)
```

```r
#중복된 데이터
plot(x, y)
```

```r
# Method 1)

#일단 분포도를 확인
xy.df <- as.data.frame(table(x, y))
xy.df
 
#중첩되는건 커져 근데 가려서 안씀.
plot(x,y,
    pch = 15,
     cex = 2 * xy.df$Freq
    )
```

```r
# Method 2)
# 이건 잘써야함

x <- c(1,1,1,2,2,2,2,2,2,3,3,4,5,6)
y <- c(1,1,4,2,3,2,2,2,3,2,1,1,1,1)

z <- data.frame(x,y)

# 중복된 갯수를 보여주는 그래프 상하좌우로 하나씩 막대기가 생김.
# 나이팅게일이 만듦...
sunflowerplot(z)
```

### chart 2개 겹쳐서 출력하기(투명도 조절)
- par(new=T) : 현재 차트를 유지하고 그 위에 다른 차트를 생성한다.
- rgb(*,*,*,*) : r(ed),g(reen),b(lue),alpha(투명도)

```r
#seq(0,1,0.1) : 0~1사이에서 0.1씩 증가시킬꺼야 로 투명도 조절도 가능!
myCol <- rgb(1,0,0,seq(0,1,0.1)) #r,g,b,alpha(투명도)


x <- 1:10
dotchart(x, col = "red")
#이사이에 두개 차트를 겹쳐서 보여줄꺼야
par(new=T) # 현재 차트를 유지하고 그 위에 다른 차트를 생성한다.
barplot(x, col = myCol)
```

### 기타그래프
#### 선분 그리기
- 선분 : 점과점사이를 잇는선
- segments(6,78,8,32, col="red")

#### 화살표 그리기
- arrows(3,56,1,12, col='blue')

#### 사각형 그리기
- rect(4,20,6,30, col = 'white', border = 'red',density = 5)

#### 글자 쓰기
- text(4,40,'This is a 샘플',srt = 50)
- str : 기울기

#### 테두리 활용하기
- mtext("character",side = 4, adj = 0.5)
- side : 각 테두리 번호
- adj 1이 제일 끝 0이 제일 앞에

#### 테두리 그리기(선)
- box(lty=1, col="red")

####  x축의 추가
- axis(1, pos=40, at=0:10,col="red")
- axis : 1 --> x축
- axis : 2 --> y축
- pos : 축 위치

#### y축의 추가
- axis(2,pos=6,at=20:60,col='blue')

```r
options(warn = -1)
par(family = "AppleGothic")

#기타그래프
x <- c(1,  3,  6,  8,  9)
y <- c(12, 56, 78, 32, 9)

plot(x,y)

#선분 그리기(3,3)과 (4,4) 선을 연결시키겠다
#선분 : 점과점사이를 잇는선
segments(6,78,8,32, col="red")

#화살표 그리기
arrows(3,56,1,12, col='blue')

#사각형 그리기
#데이터 분포 어디있을때 보여주기 편함
rect(4,20,6,30, col = 'white', border = 'red',density = 5)

#글자 쓰기
text(4,40,'This is a 샘플' ,srt = 50)

#테두리 활용하기
mtext("character",side = 4, adj = 0.5)

#chart의 테두리 그리기
box(lty=1, col="red")

#x축의 추가 : y축의 40 위치에 0~10까지 표시하고 색은 빨간색
#axis : 1 --> x//axis :2 --> y축
axis(1, pos=40, at=0:10,col="red")

#y축의 추가 : x축의 6위치에 20~60 까지 표시하고 색을 빨간색
#pos : 축 위치
axis(2,pos=6,at=20:60,col='blue')
```

```r
par(family = "AppleGothic" ,bg = "gray")
options(warn = -1)
options(scipen = 100)
barplot(pop_count[1:3,],
      ylim = c(0,500000),
       cex.names = 0.7,
       main = "투표수 높은 지역의 후보별 투표수",
        legend.text =  T,
        col = c("skyblue","red","lightgreen")
       )

for(i in 1:ncol(pop_count)){
    text(0.5+(1.2*i-1),as.numeric(pop_count[4,i])+10000,as.numeric(pop_count[4,i]), cex = 0.8)
}
```

### 수학 함수의 시각화

y= x^2
- y(tatget) = ax (Feture?)

```r
f <- function(x){
    y = x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
# 값을 많이 주니까 부드러운 곡선이 나옴
x <- seq(-5,5,0.1) # -5에서 5까지 0.1씩 증가한다.

plot(x, f(x),
    type = "l"
    )
```

y = e^(-x) -4x^2 +3|x| -3 이런 수식을 만나면

```r
f <- function(x){
    y = exp(-x) -4 * x^2 + x + 3 * abs(x) - 3
}
```

```r
plot(x,f(x),
    type = "l")
```

```r
# y = x^2 + 2x + 3
```

```r
f2 <- function(x){
     y = x^2 + 2*x + 3
}
```

```r
plot(x,f2(x),
    type = "l")
```
