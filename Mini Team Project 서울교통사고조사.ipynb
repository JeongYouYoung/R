# 년도별 교통사고 발생, 부상, 사망자수에 대한 분석

```r
seoul <- read.csv("Rdata/newSeoul2019.csv")
head(seoul)
```

```r
tail(seoul)
```

```r
str(seoul)
```

```r
summary(seoul)
```

#### 년도별 발생건수 합

```r
Accyear.sum<- aggregate(발생건수 ~ 년도 , seoul, sum)
Accyear.sum
#년도별 교통사고 발생, 부상, 사망자수에 대한 분석
options(warn = -1)
par(family = "AppleGothic")
barplot(Accyear.sum$발생건수,
        main = "년도별 교통사고 발생건수",
        ylab = "발생건수",
        ylim = c(0,50000),
        xlab = "년도",
        names = Accyear.sum$년도,
        #col = 467
       )
```

#### 년도별 부상자수 합

```r
AccInjure.sum<- aggregate(부상자수 ~ 년도 , seoul, sum)
AccInjure.sum
options(warn = -1)
par(family = "AppleGothic")
barplot(AccInjure.sum$부상자수,
        ylab = "부상자수",
        main = "년도별 교통사고 부상자수",
        ylim = c(0,60000),
        xlab = "년도",
        names = AccInjure.sum$년도
       )
```

#### 년도별 사망자수 합

```r
AccDeath.sum <- aggregate(사망자수 ~ 년도 , seoul, sum)
AccDeath.sum
options(warn = -1)
par(family = "AppleGothic")
barplot(AccDeath.sum$사망자수,
        ylab = "사망자수",
        main = "년도별 교통사고 사망자수",
        ylim = c(0,500),
        xlab = "년도",
        names = AccDeath.sum$년도
       )
```
