### female,male 선 그래프 그리기

```r
# baby 에서 여자아이만 Female이라는 dataset를 만들고 row수 구하기
head(baby)
```

```r
Female <-
subset(baby, Gender == "F")
head(Female)
nrow(Female)
```

```r
#여자아이 이름중에 중복 제거하고 갯수 구하기
#년도별로 정리했기때문에 중복이 있음.
length(unique(Female$Name))
```

```r
#유일한 여자아이 이름 20 개만 출력하기
head(unique(Female$Name),20)
```

```r
library(dplyr)
```

```r
#가장 많이 사용된 여자 아이 이름 10개 보여주기
#년도별로 이름 중복이 많으니까 중복데이터 집계 먼저

#* 1. 순위구하기
FemaleName <- aggregate(Count ~ Name, Female, sum)
head(FemaleName)
```

```r
#* 2.이름 출력
head(arrange(FemaleName, desc(Count)),10)
```

```r
# baby 에서 남자 아이만 Male이라는 데이터셋을 만들고 row수 구하기
Male <- subset(baby, Gender == "M")
nrow(Male)
```

### 년도별로 만들어서 그래프 그리기


### 구해야 할것
- 년도별 인구수

```r
head(Female)
```

```r
Female_Ycount <-
aggregate(Count ~ Year, Female, sum)
head(Female_Ycount)
```

```r
Male_Ycount <-
aggregate(Count ~ Year, Male, sum)
head(Male_Ycount)
```

```r
#Female_Ycount <- as.numeric(Female_Ycount)
```

```r
#str(Female_Ycount)
```

```r
par(family = "AppleGothic", bg="grey")
options(warn = -1)
#데이터 프레임 채로 받아올수 있는 그래프니까 그냥 데이터 프레임넣어주기
plot(Female_Ycount,
    main = "1880~2014년 인구분포 현황",
    type = "l",
    ylim = c(0,2500000),
    #xlim = c(1880,2000),
    ylab = "Count",
    xlab = "Year",
     col = "red"
    )

lines(Male_Ycount,
      type = "l",
     col = "blue")

legend(1980,1000000,
       c("Female","Male"),
       col = c("red","blue"),
       lty = c(1,1)
      )
```
