```r
봄 <- c(3,4,5)
여름 <- c(6,7,8)
가을 <- c(9,10,11)
겨울 <- c(1,2,12)

seoul <- read.csv("Rdata/newSeoul2019.csv")


        
    
   
seoul$계절 <- ifelse(seoul$월 %in% 봄,'봄',
                   ifelse(seoul$월 %in% 여름,'여름',
                          ifelse(seoul$월 %in% 가을,'가을','겨울'
                  )))
        seoul
```

```r
regionAcc.mean <- aggregate(발생건수 ~ 계절, seoul, mean)
regionAcc.mean
```

```r
regionAcc.sd <- aggregate(발생건수 ~ 계절, seoul, sd)
head(regionAcc.sd)
```

```r
regionAcc.cv <- regionAcc.sd$발생건수/regionAcc.mean$발생건수
regionAcc <- data.frame(계절=regionAcc.sd$계절,발생건변동계수=regionAcc.cv)
regionAcc$계절 <- factor(regionAcc$계절, levels=c("봄","여름","가을","겨울"))
regionAcc <- arrange(regionAcc,계절)
regionAcc
```

```r
options(scipen = 100)
options(warn = -1)
par(family="AppleGothic")
barplot(regionAcc$발생건변동계수,
       main = "계절별 발생건수_변동계수",
        xlab = "계절별",
        ylab = "변동계수",
       names = regionAcc$계절,
        cex.names=1.5,
        col = c('#FF6666','#00CCFF'),
       ylim=c(0,0.5)
       )
```
