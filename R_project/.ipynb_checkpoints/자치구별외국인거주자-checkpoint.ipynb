```r
방문객 <- c()
for(i in 1:4){
    
    for(j in 1:4){
        
        방문객 <- rbind(방문객,as.numeric(관[i,i+1])*as.numeric(외국인[,j])/100)
   
    }
    
}
방문객
ncol(방문객)
nrow(방문객)
```

```r
관광객예상 <- c()
for(i in 1:5){
    관광객예상 <- rbind(관광객예상,data.frame(예상방문객=방문객[,i]))
}
관광객예상
nrow(관광객예상)
```
