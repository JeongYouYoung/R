```r
visit <- read.csv("Rdata1/방문외국인-관광특구별상권.csv")
visit
```

```r
visit명동<-subset(visit, 관광특구 == "명동 남대문 북창동 다동 무교동")
arrange(visit명동, 년도)#desc(점포당매출금액))
boxplot(visit명동$방문외국인수,visit명동$점포당매출금액)
```

```r
visit명동<-subset(visit, 관광특구 == "명동 남대문 북창동 다동 무교동")
arrange(visit명동, 년도)

par(family="AppleGothic",bg='white'); options(warn=-1)
options(scipen=100)
plot(visit명동$방문외국인수,(visit명동$점포당매출금액*0.001)
    , main = '명동 - 방문외국인수 대비 점포당매출금액'
    , ylim = c(0,140000)
    , xlim = c(0,100000)
    , ylab = "점포당매출금액 (단위 : 1000)"
    , xlab = "방문외국인수"
    )
r <-lm((visit명동$점포당매출금액*0.001)~visit명동$방문외국인수)
abline(r, col='red')
cor(visit명동$방문외국인수,visit명동$점포당매출금액)
```

```r
방문객수합 <- c()
for(i in 1:4){
    
   방문객수합<- rbind(방문객수합,sum((subset(visit,년도==2018 & 분기 == i)[,c(2,8)])[,2]))
}

for(i in 1:4){
    
   방문객수합<- rbind(방문객수합,sum((subset(visit,년도==2019 & 분기 == i)[,c(2,8)])[,2]))
}

for(i in 1:4){
    
   방문객수합<- rbind(방문객수합,sum((subset(visit,년도==2020 & 분기 == i)[,c(2,8)])[,2]))
}

for(i in 1:4){
    
   방문객수합<- rbind(방문객수합,sum((subset(visit,년도==2021 & 분기 == i)[,c(2,8)])[,2]))
}


rownames(방문객수합) <- c('18-1','18-2','18-3','18-4',
                       '19-1','19-2','19-3','19-4',
                       '20-1','20-2','20-3','20-4',
                      '21-1','21-2','21-3','21-4'
                    )
#sum((subset(visit,년도==2018 & 분기 == 1)[,c(2,8)])[,2])
#sum((subset(visit,년도==2018 & 분기 == 2)[,c(2,8)])[,2])
```
